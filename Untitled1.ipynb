```python
# Basic data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# System and warnings
import warnings
warnings.filterwarnings("ignore")

# Preprocessing
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Time/date handling if needed
import datetime as dt

# NLP (for later text-based analysis if required)
from wordcloud import WordCloud, STOPWORDS

# Load datasets
df_employees = pd.read_csv('/content/mumbai_hr_employees.csv')
df_plan = pd.read_csv('/content/workforce_plan.csv')

# Preview data
print("üßæ Employee Dataset Sample:")
print(df_employees.head())

print("\nüìã Workforce Plan Dataset Sample:")
print(df_plan.head())

# Basic info
print("\nüìä Employee Dataset Info:")
print(df_employees.info())

print("\nüìä Workforce Plan Dataset Info:")
print(df_plan.info())
```

```python
# Copy to avoid modifying originals
employees = df_employees.copy()
plan = df_plan.copy()
```

```python
# Rename columns for ease (if needed)
employees.columns = [col.strip().lower().replace(" ", "_") for col in employees.columns]
plan.columns = [col.strip().lower().replace(" ", "_") for col in plan.columns]
```

```python
# Fill NaNs or drop rows depending on context
employees.fillna({
    'gender': 'Not Specified',
    'department': 'Unknown',
    'education': 'Not Specified',
    'city': 'Unknown',
    'recruitment_channel': 'Other',
}, inplace=True)

# For numeric columns: fill with median
num_cols = employees.select_dtypes(include=np.number).columns.tolist()
for col in num_cols:
    employees[col].fillna(employees[col].median(), inplace=True)
```

```python
# Convert date columns if available
if 'joining_date' in employees.columns:
    employees['joining_date'] = pd.to_datetime(employees['joining_date'], errors='coerce')
    employees['year_of_joining'] = employees['joining_date'].dt.year
    employees['month_of_joining'] = employees['joining_date'].dt.month

# Calculate experience if 'joining_date' exists
today = pd.to_datetime("today")
if 'joining_date' in employees.columns:
    employees['tenure_years'] = ((today - employees['joining_date']).dt.days / 365).round(1)
```

```python
cat_cols = employees.select_dtypes(include='object').columns.tolist()
le = LabelEncoder()

for col in cat_cols:
    employees[col] = le.fit_transform(employees[col])
```

```python
num_cols = employees.select_dtypes(include=np.number).columns.tolist()
```

```python
scaler = MinMaxScaler()
employees[num_cols] = scaler.fit_transform(employees[num_cols])
```

```python
# Ensure both columns are strings
employees['department'] = employees['department'].astype(str)
plan['department'] = plan['department'].astype(str)

# Now perform the merge
combined = pd.merge(employees, plan, how='left', on='department')
print("‚úÖ Successfully merged employee data with workforce planning data.")
```

```python
import ast

def safe_literal_eval(val):
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)
        except:
            return val
    return val

employees['skills'] = employees['skills'].apply(safe_literal_eval)
plan['required_skills'] = plan['required_skills'].apply(safe_literal_eval)
```

```python
# Clean up the employees['skills'] to make sure all are lists
def safe_eval(val):
    try:
        result = ast.literal_eval(val) if isinstance(val, str) else val
        return result if isinstance(result, list) else []
    except:
        return []

employees['skills'] = employees['skills'].apply(safe_eval)
```

```python
plan['required_skills'] = plan['required_skills'].apply(safe_eval)
```

```python
def is_match(emp_skills, required_skills):
    return set(required_skills).issubset(set(emp_skills))

plan['matching_employees'] = None  # initialize column

for index, row in plan.iterrows():
    required = row['required_skills']
    matched = employees[employees['skills'].apply(lambda x: is_match(x, required))]['name'].tolist()
    plan.at[index, 'matching_employees'] = matched
```

```python
print(plan[['department', 'role', 'required_skills', 'matching_employees']])
```

```python
plan['available_headcount'] = plan['matching_employees'].apply(len)
plan['headcount_status'] = plan.apply(
    lambda row: 'Sufficient' if row['available_headcount'] >= row['required_headcount'] else 'Insufficient',
    axis=1
)

print(plan[['department', 'role', 'required_headcount', 'available_headcount', 'headcount_status']])
```

```python
plan['assigned_employees'] = plan.apply(
    lambda row: row['matching_employees'][:row['required_headcount']],
    axis=1
)

print(plan[['department', 'role', 'assigned_employees']])
```

```python
used_employees = set([emp for sublist in plan['assigned_employees'] for emp in sublist])
unused_employees = employees[~employees['name'].isin(used_employees)]

print("Unused Employees:")
print(unused_employees[['name', 'skills']])
```

```python
from collections import Counter

all_skills = plan['required_skills'].sum()
skill_counts = Counter(all_skills)

skills_df = pd.DataFrame(skill_counts.items(), columns=['Skill', 'Count'])
skills_df.sort_values(by='Count', ascending=False, inplace=True)

plt.figure(figsize=(10,5))
sns.barplot(data=skills_df.head(10), x='Skill', y='Count', palette='magma')
plt.title('Top 10 In-Demand Skills')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
